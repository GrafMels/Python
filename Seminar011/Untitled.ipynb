```python
def x_xx(curent_x):
    curent_y = []
    for i in curent_x:
        curent_y.append(-12*i**4*np.sin(np.cos(i)) - 18*i**3+5*i**2 + 10*i - 30)
    return curent_y
```

```python
def xx(i):
    y = -12*i**4*np.sin(np.cos(i)) - 18*i**3+5*i**2 + 10*i - 30
    return y
```

```python
kkk = 100
k=1
d=1
list_change = []
zero_list = []
for i in np.arange(-kkk, kkk, round(0.001, 3)):
    if xx(i) > xx(i+1) and k==1:
        list_change.append(i+0.6)
        k = k * -1
    elif xx(i) < xx(i+1) and k==-1:
        list_change.append(i+0.6)
        k = k * -1
        
    if d==1 and xx(i) < 0:
        if xx(i) > xx(i-1):
            zero_list.append(i)
        else:
            zero_list.append(i)
        d = d * -1
    elif d==-1 and xx(i) > 0:
        if xx(i) > xx(i-1):
            zero_list.append(i)
        else:
            zero_list.append(i)
        d = d * -1
```

```python
z = 1
plt.figure(figsize=(10, 5))
for i in range(len(list_change)-1):
    curent_x = np.arange(list_change[i], list_change[i+1], 0.001)
    curent_y = x_xx(curent_x)
    if z == 1:
        plt.plot(curent_x, curent_y, 'r')
    elif z == -1:
        plt.plot(curent_x, curent_y, 'b')
    z = z * -1 
ax = plt.gca()

# plot X - axis    
ax.axhline(y=0, color='k')

# plot Y - axis    
ax.axvline(x=0, color='k')
for i in range(len(zero_list)):
    plt.scatter(zero_list[i], 0, color='orange', s=20, marker='o')

plt.show()
```
